```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset from the provided path
file_path = 'pubg-weapon-stats.csv'

# Load the uploaded file into a pandas DataFrame
df = pd.read_csv(file_path)

# Display the first few rows of the DataFrame
print(df.head())
```

```python
print(df.info())
```

```python
print(df.describe())
```

# 1. Weapon Type Distribution

```python
plt.figure(figsize=(10, 6))
df['Weapon Type'].value_counts().plot(kind='barh')
plt.title('Distribution of Weapon Types')
plt.xlabel('Count')
plt.ylabel('Weapon Type')
plt.show()

```

# 2. Bullet Type Distribution


```python
plt.figure(figsize=(10, 6))
df['Bullet Type'].value_counts().plot(kind='barh')
plt.title('Distribution of Bullet Types')
plt.xlabel('Count')
plt.ylabel('Bullet Type')
plt.show()

```

# 3. Average Damage by Weapon Type

```python
plt.figure(figsize=(10, 6))
df.groupby('Weapon Type')['Damage'].mean().sort_values().plot(kind='barh')
plt.title('Average Damage by Weapon Type')
plt.xlabel('Average Damage')
plt.ylabel('Weapon Type')
plt.show()

```

# 4. Average Range by Weapon Type

```python
plt.figure(figsize=(10, 6))
df.groupby('Weapon Type')['Range'].mean().sort_values().plot(kind='barh')
plt.title('Average Range by Weapon Type')
plt.xlabel('Average Range')
plt.ylabel('Weapon Type')
plt.show()

```

# 5. Average Damage by Bullet Type

```python
plt.figure(figsize=(10, 6))
df.groupby('Bullet Type')['Damage'].mean().sort_values().plot(kind='barh')
plt.title('Average Damage by Bullet Type')
plt.xlabel('Average Damage')
plt.ylabel('Bullet Type')
plt.show()

```

# 6. Average Range by Bullet Type

```python
plt.figure(figsize=(10, 6))
df.groupby('Bullet Type')['Range'].mean().sort_values().plot(kind='barh')
plt.title('Average Range by Bullet Type')
plt.xlabel('Average Range')
plt.ylabel('Bullet Type')
plt.show()

```

# 7. Calculate The Average Range For Each Weapon Type

```python
average_range_by_weapon_type = df.groupby('Weapon Type')['Range'].mean().reset_index()
print(average_range_by_weapon_type)

```

```python
import matplotlib.pyplot as plt

# Assuming 'average_range_by_weapon_type' DataFrame is already defined

# Plotting the data
plt.figure(figsize=(10, 6))  # Adjust the figure size as needed
plt.bar(average_range_by_weapon_type['Weapon Type'], average_range_by_weapon_type['Range'], color='skyblue')
plt.xlabel('Weapon Type')
plt.ylabel('Average Range')
plt.title('Average Range by Weapon Type')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability if necessary
plt.grid(True)
plt.tight_layout()

# Show plot
plt.show()
```
